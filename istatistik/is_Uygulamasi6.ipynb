# **İş Uygulaması: Web Sitesinde Geçirilen Sürenin Testi**

#### Problem: Web sitemizde geçirilen ortalama süre gerçekten 170 saniye mi?
#### Detaylar: Yazılımlardan elde edilen web sitesinde geçirilen ort. süreler var. Bu veriler incelendiğinde bir yönetici ya da çalışanımız bu değerlerin böyle olmadığına yönelik düşünceler taşıyor ve bu durumu test etmek istiyorlar.

#### **Hipotezler:**
#### H0; mü=50
#### H1; mü!=50
#### H0; mü<=50
#### H1; mü>50
#### H0; mü>=50
#### H1; mü<50
#### **Test İstatistiği:**
#### t=(ortalama-mü0)/(s/kökn)

```python
import numpy as np
import pandas as pd
olcumler = np.array([17,160,234,149,145,107,197,75,201,225,211,119,
                     157,145,127,244,163,114,145,65,112,185,202,146,
                     203,224,203,114,188,156,177,95,165,50,110,216,138,
                     151,166,135,155,84,251,173,131,207,121,120])
```

```python
olcumler[0:10]
```

```python
import scipy.stats as stats
```

```python
stats.describe(olcumler)
```

```python
#varsayımlar
#normallik varsayımı
```

```python
#histogram
pd.DataFrame(olcumler).plot.hist();
```

```python
#qqplot
import pylab
stats.probplot(olcumler,dist="norm",plot=pylab)
pylab.show()
```

```python
#Shapiro-Wilks Testi
```

#### H0: Örnek dağılımı ile teorik normal dağılım arasında ist. ol. anl. bir fark. yoktur
#### H1: ...Fark vardır

```python
from scipy.stats import shapiro
```

```python
shapiro(olcumler)
```

```python
print("T Hesap İstatistiği: "+str(shapiro(olcumler)[0]))
print("Hesaplanan P-value: "+str(shapiro(olcumler)[1]))
```

```python
#Hipotez testinin uygulanması
```

```python
stats.ttest_1samp(olcumler, popmean=170)
```

H0: web sitemizde geçirilen ortalama süre 170'tir
H1: ..değiştir

## **Nonparametrik Tek Örneklem Testi**

```python
from statsmodels.stats.descriptivestats import sign_test
```

```python
sign_test(olcumler,170)
```

### **Tek Örneklem Oran Testi**
#### Oransal bir ifade test edilmek istenildiğinde kullanılır.
